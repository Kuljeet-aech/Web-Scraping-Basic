```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
url = "https://dharani.telangana.gov.in/knowLandStatus"
driver.get(url)

# Select the 'District' dropdown
district_dropdown = driver.find_element(By.ID,value="districtID")
district_select = Select(district_dropdown)

# Extract all options in the 'District' dropdown
district_options = [option.text for option in district_select.options]

# Print all the options in District Dropdown
print("Select a district:")
for i, district in enumerate(district_options):
    print(f"{i}. {district}")
```

```python
# Prompt to input the index of the 'District' option
selected_index_district = int(input("Enter the index of the district option you want to select: "))

# Select the 'District' option from the prompt
district_select.select_by_index(selected_index_district)
```

```python
# Find the 'Mandal' dropdown after selection
mandal_dropdown = driver.find_element(By.ID,value="mandalID")
mandal_select = Select(mandal_dropdown)

# Extract all options in the 'Mandal' dropdown
mandal_options = [option.text.strip() for option in mandal_select.options]

# Print the options for 'Mandal' dropdown
print("\nMandal Options after selecting", district_select.first_selected_option.text.strip() + ":")
for i, mandal in enumerate(mandal_options):
    print(f"{i}. {mandal}")
```

```python
# Prompt to input the index of the Mandal option
selected_index_mandal = int(input("Enter the index of the Mandal option you want to select: "))

# Select the 'Mandal' option
mandal_select.select_by_index(selected_index_mandal)
```

```python
# Find the 'Village' dropdown after selection
village_dropdown = driver.find_element(By.ID,value="villageId")
village_select = Select(village_dropdown)

# Extract all options in the 'Village' dropdown
village_options = [option.text.strip() for option in village_select.options]

# Print the options of 'Village' dropdown
print("\nVillage Options after selecting", mandal_select.first_selected_option.text.strip() + ":")
for i, village in enumerate(village_options):
    print(f"{i}. {village}")
```

```python
# Prompt to input the index of the Village option
selected_index_village = int(input("Enter the index of the Village option you want to select: "))

# Select the 'Village' option
village_select.select_by_index(selected_index_village)
```

```python
# Find the 'Survey No.' dropdown after selection
Survey_dropdown = driver.find_element(By.ID,value="surveyIdselect")
Survey_select = Select(Survey_dropdown)

# Extract all options in the 'Survey No.' dropdown
Survey_options = [option.text.strip() for option in Survey_select.options]

# Print all the final required options 
print("\nSurvey Options after selecting", village_select.first_selected_option.text.strip() + ":")
for i, survey in enumerate(Survey_options):
    print(f"{i}. {survey}")
```
